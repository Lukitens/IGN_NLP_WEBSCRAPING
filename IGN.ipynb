```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("resultIGN.json")
df
```

```python
df.dropna(subset="contenido", inplace=True)
df
```

```python
df = df.explode("titulo")
df.drop("fecha_de_publicacion", axis=1, inplace=True)
df
```

```python
df["contenido"] = df["contenido"].astype(str)

df["contenido"] = df["contenido"].str.replace("[", "")
df["contenido"] = df["contenido"].str.replace("]", "")
df["contenido"] = df["contenido"].str.replace("'", "")
df["contenido"] = df["contenido"].str.replace('"', "")
df
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(inplace=True, drop=True)
df
```

```python
import nltk
from nltk.tokenize import word_tokenize

#Creamos la lista de stopwords en español
stopwords = nltk.corpus.stopwords.words("spanish")

#Creamos la funcion para remover las stopwords

df["contenido"] = df["contenido"].str.lower()

def remover_stopwords(texto):

    tokens = word_tokenize(texto, language='spanish')

    texto_sin_stopwords = [palabra for palabra in tokens if palabra not in stopwords]

    return texto_sin_stopwords

df_nlp = df
df_nlp["contenido_sin_stopwords"] = df_nlp["contenido"].apply(lambda x : remover_stopwords(x))
df_nlp
```

```python
df_nlp["contenido_sin_stopwords"] = df_nlp["contenido_sin_stopwords"].apply(lambda x: ' '.join(x))
df_nlp["contenido_sin_stopwords"] = df_nlp["contenido_sin_stopwords"].str.replace(", ,", "")
df_nlp["contenido_sin_stopwords"] = df_nlp["contenido_sin_stopwords"].str.replace(",,", "")
df_nlp
```

```python
# Importar el tokenizador específico para el español
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')

from nltk.stem import WordNetLemmatizer
lematizer = WordNetLemmatizer()

def lematizador(texto):

    tokens = tokenizer.tokenize(texto)

    lematizado = []

    for i in tokens:
        lematizado.append(lematizer.lemmatize(i, pos='v'))

    return lematizado

df_nlp["contenido_sin_stopwords"] = df_nlp["contenido_sin_stopwords"].apply(lambda x: lematizador(x))
df_nlp["contenido_sin_stopwords"] = df_nlp["contenido_sin_stopwords"].apply(lambda x: ' '.join(x))
df_nlp["contenido_sin_stopwords"]
```

```python
from gensim.models import KeyedVectors
import gensim.downloader as api

wv = api.load('fasttext-wiki-news-subwords-300')
```

```python
def reescribir_texto(texto, modelo):
    palabras = tokenizer.tokenize(texto)
    texto_reescrito = ""
    for palabra in palabras:
        if palabra in modelo.key_to_index:
            sinonimo = modelo.most_similar(palabra)[0][0]
            texto_reescrito += sinonimo + " "
        else:
            texto_reescrito += palabra + " "  # Si la palabra no está en el modelo, mantenerla sin cambios
    return texto_reescrito.strip()  # Eliminar el espacio en blanco extra al final

df["sinonimos"] = df_nlp["contenido_sin_stopwords"].apply(lambda x: reescribir_texto(x, wv))
df["sinonimos"]
```
